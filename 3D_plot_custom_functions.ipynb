Lets concider few other commonly used matrics
1. squared_error variations
2. absolute_error

```
def squared_error(a,p):
  return (a-p)**2

def absolute_error(a,p):
  return abs(a-p)
```

```
fig1 = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.contour3D(m, n, squared_error(m, n), 100, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z');
```

Change the angle of View

```
ax1.view_init(10, 40)
fig1
```

```
fig1 = plt.figure()
ax1 = plt.axes(projection='3d')
ax1.contour3D(m, n, absolute_error(m, n), 100, cmap='viridis')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z');
```
